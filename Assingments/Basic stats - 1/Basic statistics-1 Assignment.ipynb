Descriptive Analytics for Numerical Columns
===========================================

Mean
----
 * Mean defines average value of the data set.
 * We have two kind of mean. One is population mean and another one is sample mean.
 * Population mean is taken from ( Sum of whole data/ count of data).
 * Sample mean is taken from (Sum of Sample data/ count of data).
   
Median
------
 * Median is defines centrel value of the data set.
 * Before taking median we need to arrange the data in ascending order and then take mid value of the dataset.
 * If count of the dataset is even take middle two value and divide by 2.
   
Mode
----
 * Mode define hight no of repeated value in the dataset.

Standard Deviation
------------------
 * Standard Deviation define spread of the dataset from the mean.
 * It is telling that distance of each value from the mean.
 * If the standard deviation is high means dataspread also high.
 * If Standard deviation is low means dataspread also low.
 * We have two kind of Standard Deviation one is Population Standard Deviation and another one is Sample Standard  Deviation.
 * Population Standard Deviation = sqrt(1\N.[(x1-μ)^2+(x2-μ)^2+.....+(xi-u)^2]).
 * μ - population mean.
 * Sample Standard Deviation = sqrt(1\1-n.[(x1-x_bar)^2+(x2-x_bar)^2+.....+(xi-x_bar)^2]).
 * x-bar -Sample mean.

Standardization of Numerical Variables

Z-score normalization 

* Used to find the divation of the datapoint from the mean of the dataset.
* We can use Z-score distribution for both sample and population.
* For population Z = (x-mu)\sigma.
* For sample Z = (x_bar-mu)\sigma\sqrt(n).
* Positive Z-score value is indicating that the datapoint is above the mean.
* Negative Z-score value is indicating that the datapoint is below the mean.

```python
#We can simply find Z-score for all numerical values in the table like belowe.

Data['Zscore_Volume'] =(Data['Volume']-Data['Volume'].mean())/Data['Volume'].std()
Data['Zscore_Avg_Price'] =(Data['Avg_Price']-Data['Avg_Price'].mean())/Data['Avg_Price'].std()
Data['Zscore_Total_Sales_Value'] =(Data['Total_Sales_Value']-Data['Total_Sales_Value'].mean())/Data['Total_Sales_Value'].std()
Data['Zscore_Discount_Rate'] =(Data['Discount_Rate']-Data['Discount_Rate'].mean())/Data['Discount_Rate'].std()
Data['Zscore_Discount_Amount'] =(Data['Discount_Amount']-Data['Discount_Amount'].mean())/Data['Discount_Amount'].std()
Data['Zscore_Net_Sales_Value'] =(Data['Net_Sales_Value']-Data['Net_Sales_Value'].mean())/Data['Net_Sales_Value'].std()
Data


#In the above coding i have adding new column for each numerical columns Z-scores values.
#Let me explain step by step
#Assinging the New column value in Data table as Zscore_Volume.
#Taking each value from the volume column.
#Taking mean for each value in volume column.
#Taking std for each value in volume column.
#Calculating Zscore value using the above taken values.
```
