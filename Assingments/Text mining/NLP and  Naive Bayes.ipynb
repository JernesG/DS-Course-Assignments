```python
#Data cleaning function to clean each sentence separately
import re 
import string

def clean_Data(Data):
    
    Data = Data.lower()
    Data = re.sub('\[.*?\]', '', Data)
    Data = re.sub('[%s]' % re.escape(string.punctuation), '', Data)
    Data = re.sub('\w*\d\w*', '', Data)
    Data = re.sub("[0-9" "]+"," ",Data)
    Data = re.sub('[‚Äò‚Äô‚Äú‚Äù‚Ä¶]', '', Data)
    Data = re.sub('üòç','',Data)
    return Data  

clean = lambda x: clean_Data(x)
```

```python
#Calling the cleaning function and clean the sentence
Data1['Data'] = Data1["Data"].apply(clean)
Data.Data
```

```python
#Spliting the dataset into Train and Test for model training and testing
from sklearn.model_selection import train_test_split

blogs_categories_train,blogs_categories_test = train_test_split(Data,test_size=0.2)
```

```python
# Word count matrix format for classification 
from sklearn.feature_extraction.text import CountVectorizer
blogs_categories_bow = CountVectorizer().fit(Data1.Data)
```

```python
# For all messages
all_blogs_categories_matrix = blogs_categories_bow.transform(Data1.Data)
all_blogs_categories_matrix.shape 
```

```python
X1 = all_blogs_categories_matrix.toarray()
```

```python
df = pd.DataFrame(X1, columns=blogs_categories_bow.get_feature_names_out())
df
```

```python
# For training messages
train_blogs_categories_matrix = blogs_categories_bow.transform(blogs_categories_train.Data)
train_blogs_categories_matrix.shape

# For testing messages
test_blogs_categories_matrix = blogs_categories_bow.transform(blogs_categories_test.Data)
test_blogs_categories_matrix.shape 
```

```python
train_blogs_categories_matrix.toarray()
```

```python
# Preparing a naive bayes model on training data set 
import numpy as np
from sklearn.naive_bayes import MultinomialNB as MB
from sklearn.metrics import classification_report


# Multinomial Naive Bayes
classifier_mb = MB()
classifier_mb.fit(train_blogs_categories_matrix,blogs_categories_train.Labels)

train_pred_m = classifier_mb.predict(train_blogs_categories_matrix)
accuracy_train_m = np.mean(train_pred_m==blogs_categories_train.Labels) 

print(classification_report(train_pred_m,blogs_categories_train.Labels))

test_pred_m = classifier_mb.predict(test_blogs_categories_matrix)
accuracy_test_m = np.mean(test_pred_m==blogs_categories_test.Labels)
```
