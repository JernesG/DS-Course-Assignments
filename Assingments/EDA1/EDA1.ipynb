```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Reading data
data = pd.read_csv("Cardiotocographic.csv")
data
```

```python
#By using the below comment we can see and verify the datatypes of the data
data.info()
```

```python
#Identifying the missing values in the data

data.isnull().sum()
```

```python
#Finding median for each columns to fill null values

Median_LB = data["LB"].median()
Median_AC = data["AC"].median()
Median_DS = data["DS"].median()
Median_DP = data["DP"].median()
Median_MLTV = data["MLTV"].median()
Median_Width = data["Width"].median()
Median_Tendency = data["Tendency"].median()
Median_NSP = data["NSP"].median()


#Fill column_median values in the null values on the each columns

data["LB"] = data["LB"].fillna(Median_LB)
data["AC"] = data["AC"].fillna(Median_AC)
data["DS"] = data["DS"].fillna(Median_DS)
data["DP"] = data["DP"].fillna(Median_DP)
data["MLTV"] = data["MLTV"].fillna(Median_MLTV)
data["Width"] = data["LB"].fillna(Median_Width)
data["Tendency"] = data["LB"].fillna(Median_Tendency)
data["NSP"] = data["LB"].fillna(Median_NSP)

#Checking again the null count on each row to verify whether the median value is filled or not
data.isnull().sum()
```

```python
#Checking for duplicates in the table

data[data.duplicated(keep= False)]
```

```python
#Droping duplicate rows
data = data.drop_duplicates(keep = 'first')
```

```python
#Checking again is there any duplicate rows are present
data[data.duplicated(keep= False)]
```

```python
#Detecting Outliers

#We can find the outliers by Many methods like Z-score values,IQR methods and Visualization

#IQR methods
# Calculate Q1 and Q3
Q1_LB=np.percentile((data["LB"]),25)
Q3_LB=np.percentile((data["LB"]),75)

# Calculate IQR
IQR = Q3_LB - Q1_LB

# Determine outlier boundaries
lower_bound = Q1_LB - 1.5 * IQR
upper_bound = Q3_LB + 1.5 * IQR

print("Q1:", Q1_LB)
print("Q3:", Q3_LB)
print("IQR:", IQR)
print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)

# Identify outliers
for x in data["LB"]:
    if (x<lower_bound):
        print("Lower Outliers:",x)
    elif (x>upper_bound):
        print("Upper Outliers:",x)
```
